```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args

    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)

    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
def length_category(mins):
    if mins < 2:
        return 'Short'
    elif mins < 4:
        return 'Medium'
    elif mins < 6:
        return 'Long'
    else:
        return 'Epic'

spotify['length_category'] = spotify['duration_min'].apply(length_category)
```

```python
spotify.columns
```

```python
spotify['length_category']
```

```python
num_samples = 5

short_tracks = spotify[spotify['length_category'] == 'Short'].sample(num_samples)
medium_tracks = spotify[spotify['length_category'] == 'Medium'].sample(num_samples)
long_tracks = spotify[spotify['length_category'] == 'Long'].sample(num_samples)
epic_tracks = spotify[spotify['length_category'] == 'Epic'].sample(num_samples)
```

```python
playlist_dataframe = pd.concat([short_tracks, medium_tracks, long_tracks, epic_tracks])
playlist_dataframe[['track_name', 'artists', 'duration_min', 'length_category']]
```

```python
def mood(row):
    if row['energy'] > 0.7 and row['valence'] > 0.6:
        return 'Hype'
    elif row['energy'] < 0.4 and row['acousticness'] > 0.5:
        return 'Chill'
    elif row['valence'] < 0.3:
        return 'Dark'
    elif row['danceability'] > 0.7:
        return 'Dancey'
    else:
        return 'Mixed'

spotify['mood'] = spotify.apply(mood, axis=1)
```

```python

top_tracks = spotify.sort_values('popularity', ascending=False).head(20)

# Set up figure
plt.figure(figsize=(14, 8))
ax = plt.subplot()

colors = plt.cm.Greens(np.linspace(0.4, 1, len(top_tracks)))

# Create horizontal bar plot
bars = ax.barh(top_tracks['track_name'].str[:30] + "...",  # Truncate long names
              top_tracks['popularity'],
              color=colors,
              edgecolor='black',
              height=0.7)

# Add popularity values on bars
ax.bar_label(bars,
             fmt='%d',
             padding=5,
             fontsize=9,
             color='black')

# Customize appearance
ax.set_title('Top 20 Most Popular Tracks On Spotify',
             fontsize=16,
             pad=20,
             fontweight='bold')
ax.set_xlabel('Popularity',
              fontsize=12)
ax.set_ylabel('Track Name',
              fontsize=12)
ax.set_xlim(0, 100)

plt.text(0.95, 0.05, 'Spotify',
         transform=ax.transAxes,
         fontsize=40,
         color='lightgray',
         alpha=0.3,
         ha='right',
         va='bottom')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)


ax.invert_yaxis()

plt.tight_layout()
plt.show()
```
